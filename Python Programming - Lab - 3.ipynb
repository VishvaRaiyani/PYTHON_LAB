### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input("Enter a Year : "))
if((year % 4 == 0 and year % 100 != 0) or year % 400 == 0) :
    print(f"{year} is Leap Year.")
else:
    print(f"{year} is not Leap Year.")
```

### 06) WAP in python to display the name of the day according to the number given by the user.

```python
day = int(input("Enter a number"));

match day:
    case 1:
        print("Monday.");
    case 2:
        print("Tuesday.");
    case 3:
        print("Wednesday.");
    case 4:
        print("Thursday.");
    case 5:
        print("Friday.");
    case 6:
        print("Saturday.");
    case 7:
        print("Sunday.");
    case _:
        print("Invalid.");
```

### 07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
a , b , op = input("Enter a Number and Operator : ").split(" ");
a = int(a)
b = int(b)

match op:
    case '+':
        print(f"Sum = {a + b}");
    case '-':
        print(f"Subtract = {a - b}");
    case '*':
        print(f"Multiplication = {a * b}");
    case '/':
        if( b != 0) :
            print(f"Division = {a / b}");
    case _:
        print("Invalid")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
a , b , c , d , e = input("Enter marks of five subjects.").split(" ")
a = int(a)
b = int(b)
c = int(c)
d = int(d)
e = int(e)

sum = a + b + c + d + e

avg = float(sum / 5);

if(avg < 35):
    print("Fail!!.");
elif(avg >= 35 and avg < 45):
    print("Pass Class.");
elif(avg >= 45 and avg < 60):
    print("Second Class.")
elif(avg >= 60 and avg < 70):
    print("First Class.")
else :
    print("Distinction.")

```

### 10) WAP to calculate electricity bill based on following criteria. Which takes the unit from the user.
#### a. First 1 to 50 units – Rs. 2.60/unit
#### b. Next 50 to 100 units – Rs. 3.25/unit
#### c. Next 100 to 200 units – Rs. 5.26/unit
#### d. above 200 units – Rs. 8.45/unit

```python
unit = int(input("Enter Unit :"))
if(unit >= 1 and unit <= 50) :
    print(f"Price = {unit * 2.60}")
elif(unit <= 100) :
    print(f"Price = {50 * 2.60 + (unit - 50) * 3.25}")
elif(unit <= 200) :
    print(f"Price = {50 * 2.60 + (unit - 50) * 3.25 + (unit - 100) * 5.26}")
else : 
    print(f"Price = {50 * 2.60 + (unit - 50) * 3.25 + (unit - 100) * 5.26 + (unit - 200) * 8.45} ")
```
