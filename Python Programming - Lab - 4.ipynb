### 13) WAP to print sum of digits of given number.

```python
a = int(input("Please Enter a number."))
temp = a
sum = 0;
while a > 0:
    rem = a % 10;
    a = a // 10
    sum += rem
print(f"Sum of digit({temp}) = {sum}")
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
a = int(input("Please Enter a number."))
sum = 0;
temp = a;
while a > 0:
    rem = a % 10;
    sum = sum * 10 + rem;
    a = a // 10

if(temp == sum):
    print(f"{temp} is palindrom")
else:
    print(f"{temp} is not palindrom")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
a = int(input("Please Enter a number."))
sum = 0;
temp = a;
power = len(str(a))
while a > 0:
    rem = a % 10;
    a = a // 10
    sum = sum + (rem ** power)
if(temp == sum):
    print(f"{temp} is Armstrong.")
else:
    print(f"{temp} is not Armstrong.")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
a = int(input("Please Enter a number."))
for i in range(1 , a+1):
    sum = 0
    for j in range(1 , i):
        if(i % j == 0):
            sum = sum + j;
    if(sum == i):
        print(i)
```
